```python

from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.docstore.document import Document
import pandas as pd
import os


DATA_PATH = "/content/drive/MyDrive/farmlingua/data/raw"
VECTOR_DB_PATH = "/content/drive/MyDrive/farmlingua/vectorstore/faiss_index"


from google.colab import drive
drive.mount('/content/drive')


pdf_folder = os.path.join(DATA_PATH, "knowledge")
pdf_docs = []

for filename in os.listdir(pdf_folder):
    if filename.endswith(".pdf"):
        try:
            loader = PyPDFLoader(os.path.join(pdf_folder, filename))
            pdf_docs.extend(loader.load())
            print(f" Loaded PDF: {filename}")
        except Exception as e:
            print(f" Failed to load {filename}: {str(e)}")

print(f"\n Successfully loaded {len(pdf_docs)} documents from PDFs.")


csv_folder = os.path.join(DATA_PATH, "farmer_queries")
csv_docs = []

ENCODINGS_TO_TRY = ['utf-8', 'latin1', 'ISO-8859-1', 'cp1252']

for filename in os.listdir(csv_folder):
    if filename.endswith(".csv"):
        file_path = os.path.join(csv_folder, filename)
        loaded = False


        for encoding in ENCODINGS_TO_TRY:
            try:

                loader = CSVLoader(file_path=file_path, encoding=encoding)
                docs = loader.load()
                csv_docs.extend(docs)
                print(f" Loaded CSV ({encoding}): {filename}")
                loaded = True
                break
            except UnicodeDecodeError:
                continue
            except Exception as e:
                print(f" CSVLoader failed for {filename} with {encoding}: {str(e)}")
                continue


        if not loaded:
            try:
                df = pd.read_csv(file_path, encoding='latin1')
                docs = [Document(page_content=str(row), metadata={"source": filename})
                       for row in df.to_dict(orient='records')]
                csv_docs.extend(docs)
                print(f" Loaded CSV (pandas fallback): {filename}")
            except Exception as e:
                print(f" Failed to load {filename} after all attempts: {str(e)}")

print(f"\n Successfully loaded {len(csv_docs)} rows from CSVs.")

if not (pdf_docs or csv_docs):
    raise ValueError("No documents were loaded. Check your input files.")

all_docs = pdf_docs + csv_docs

splitter = RecursiveCharacterTextSplitter(
    chunk_size=512,
    chunk_overlap=64,
    length_function=len,
    add_start_index=True
)


split_docs = splitter.split_documents(all_docs)

print(f"\n Total chunks created: {len(split_docs)}")


embedding_model = SentenceTransformerEmbeddings(
    model_name="sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2"
)

try:
    vectorstore = FAISS.from_documents(split_docs, embedding_model)
    print("\n Vectorstore created successfully")
except Exception as e:
    print(f"\n Failed to create vectorstore: {str(e)}")
    raise


try:
    os.makedirs(VECTOR_DB_PATH, exist_ok=True)
    vectorstore.save_local(VECTOR_DB_PATH)
    print(f"\n Vectorstore saved to: {VECTOR_DB_PATH}")


    if os.path.exists(os.path.join(VECTOR_DB_PATH, "index.faiss")):
        print(" Verification: FAISS files found successfully")
    else:
        print(" Warning: FAISS files not found after save")
except Exception as e:
    print(f"\n Failed to save vectorstore: {str(e)}")
    raise

print("\n Pipeline completed successfully ")
```

THE BUILDING OF THE INTENT CLASSIFIER

```python
import pandas as pd
import numpy as np
import os
import chardet
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.feature_extraction.text import TfidfVectorizer
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, balanced_accuracy_score
from sklearn.utils.class_weight import compute_sample_weight
import joblib
from collections import Counter
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline
import warnings


warnings.filterwarnings('ignore')

DATA_PATH = "/content/drive/MyDrive/farmlingua/data/raw/farmer_queries"
MODEL_PATH = "/content/drive/MyDrive/farmlingua/models/intent_classifier_v2.joblib"

def robust_csv_loader(file_path):
    """Advanced CSV loader with multiple fallback strategies"""
    encodings = ['utf-8', 'utf-8-sig', 'latin1', 'ISO-8859-1', 'cp1252', 'windows-1250']
    delimiters = [',', ';', '\t']

    for encoding in encodings:
        for delimiter in delimiters:
            try:
                df = pd.read_csv(file_path, encoding=encoding, delimiter=delimiter)
                if df.shape[1] >= 2:

                    text_cols = [c for c in df.columns if 'question' in c.lower() or 'text' in c.lower()]
                    label_cols = [c for c in df.columns if 'intent' in c.lower() or 'label' in c.lower()]

                    if text_cols and label_cols:
                        return df[[text_cols[0], label_cols[0]]].rename(columns={
                            text_cols[0]: 'English_Question',
                            label_cols[0]: 'Intent'
                        })
            except:
                continue


    try:
        with open(file_path, 'rb') as f:
            content = f.read().decode('latin1', errors='ignore')
        lines = [line.strip().split(',') for line in content.split('\n') if line.strip()]
        if len(lines) > 1:
            return pd.DataFrame(lines[1:], columns=['English_Question', 'Intent'])
    except Exception as e:
        print(f"Could not load {file_path}: {str(e)}")
        return None


dfs = []
for filename in os.listdir(DATA_PATH):
    if filename.endswith(".csv"):
        file_path = os.path.join(DATA_PATH, filename)
        print(f"\nLoading: {filename}")

        df = robust_csv_loader(file_path)
        if df is not None:

            df = df.dropna().copy()
            df['English_Question'] = df['English_Question'].astype(str).str.strip()
            df['Intent'] = df['Intent'].astype(str).str.strip()


            df = df[(df['English_Question'] != '') & (df['Intent'] != '')]

            if not df.empty:
                dfs.append(df)
                print(f" Success: Loaded {len(df)} samples")
            else:
                print(f" Warning: No valid data in {filename}")
        else:
            print(f" Failed: Could not load {filename}")

if not dfs:
    raise ValueError("No CSV files could be loaded. Please check your data files.")

all_data = pd.concat(dfs, ignore_index=True)


print("\nClass distribution BEFORE balancing:")
print(all_data['Intent'].value_counts())


class_counts = all_data['Intent'].value_counts()
valid_classes = class_counts[class_counts >= 5].index
all_data = all_data[all_data['Intent'].isin(valid_classes)]


le = LabelEncoder()
all_data['Intent_encoded'] = le.fit_transform(all_data['Intent'])


X_train, X_test, y_train, y_test = train_test_split(
    all_data["English_Question"],
    all_data['Intent_encoded'],
    test_size=0.2,
    random_state=42,
    stratify=all_data['Intent_encoded']
)

X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train,
    test_size=0.25,
    random_state=42,
    stratify=y_train
)


vectorizer = TfidfVectorizer(
    ngram_range=(1, 2),
    max_features=10000,
    stop_words='english',
    min_df=2,
    max_df=0.95
)


X_train_tfidf = vectorizer.fit_transform(X_train)
smote = SMOTE(sampling_strategy='not majority', random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train_tfidf, y_train)


base_clf = XGBClassifier(
    objective='multi:softprob',
    eval_metric='mlogloss',
    num_class=len(le.classes_),
    early_stopping_rounds=20,
    n_estimators=500,
    tree_method='hist',
    enable_categorical=False
)

param_grid = {
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0],
    'gamma': [0, 0.1],
    'reg_alpha': [0, 0.1],
    'reg_lambda': [1, 0.1]
}


def balanced_f1(y_true, y_pred):
    return balanced_accuracy_score(y_true, y_pred)

grid_search = GridSearchCV(
    estimator=base_clf,
    param_grid=param_grid,
    cv=StratifiedKFold(n_splits=3),
    scoring=balanced_f1,
    n_jobs=-1,
    verbose=2
)

print("\nStarting hyperparameter tuning...")
grid_search.fit(
    X_train_res,
    y_train_res,
    eval_set=[(vectorizer.transform(X_val), y_val)],
    verbose=True
)

best_clf = grid_search.best_estimator_
print(f"\nBest parameters: {grid_search.best_params_}")


def evaluate_model(model, X, y_true, classes):
    y_pred = model.predict(X)
    y_proba = model.predict_proba(X)

    print("\nClassification Report:")
    print(classification_report(y_true, y_pred, target_names=classes, zero_division=0))

    print(f"\nBalanced Accuracy: {balanced_accuracy_score(y_true, y_pred):.2f}")


    class_acc = {}
    for c in np.unique(y_true):
        mask = y_true == c
        class_acc[classes[c]] = (y_pred[mask] == y_true[mask]).mean()

    print("\n10 Worst Performing Classes:")
    print(pd.Series(class_acc).sort_values().head(10))

    return y_pred, y_proba

print("\n=== Validation Set Performance ===")
y_val_pred, y_val_proba = evaluate_model(best_clf, vectorizer.transform(X_val), y_val, le.classes_)

print("\n=== Test Set Performance ===")
y_test_pred, y_test_proba = evaluate_model(best_clf, vectorizer.transform(X_test), y_test, le.classes_)


feature_names = vectorizer.get_feature_names_out()
importance = best_clf.feature_importances_
top_features = pd.DataFrame({
    'feature': feature_names,
    'importance': importance
}).sort_values('importance', ascending=False).head(20)

print("\nTop 20 Important Features:")
print(top_features)


joblib.dump(
    {
        'model': best_clf,
        'vectorizer': vectorizer,
        'label_encoder': le,
        'smote': smote,
        'feature_importance': top_features,
        'classes': le.classes_,
        'class_distribution': pd.Series(y_train_res).value_counts()
    },
    MODEL_PATH
)
print(f"\nModel saved to: {MODEL_PATH}")


def predict_intent(text, model_path=MODEL_PATH, confidence_threshold=0.6):
    """Enhanced prediction with confidence thresholding"""
    artifacts = joblib.load(model_path)


    text_processed = str(text).strip()
    if not text_processed:
        return "INVALID_INPUT", 0.0


    vec = artifacts['vectorizer'].transform([text_processed])


    proba = artifacts['model'].predict_proba(vec)[0]
    predicted_idx = np.argmax(proba)
    confidence = proba[predicted_idx]


    if confidence < confidence_threshold:
        top3_idx = np.argsort(proba)[-3:][::-1]
        top3 = [(artifacts['classes'][i], proba[i]) for i in top3_idx]
        return f"LOW_CONFIDENCE_TOP3: {top3}", confidence

    return artifacts['classes'][predicted_idx], confidence

sample_queries = [
    "How to control cassava pests?",
    "Best fertilizer for maize",
    "This is a random unrelated question"
]

print("\nSample Predictions:")
for query in sample_queries:
    pred, conf = predict_intent(query)
    print(f"'{query}'  {pred} (confidence: {conf:.2f})")
```

CREW ai agent structure

```python

from crewai import Agent, Crew, Task
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.llms import HuggingFacePipeline
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
from joblib import load
import torch


intent_bundle = load('/content/drive/MyDrive/farmlingua/models/intent_classifier_v2.joblib')
intent_model = intent_bundle['model']
vectorizer = intent_bundle['vectorizer']
label_encoder = intent_bundle['label_encoder']


embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2"
)

live_rag_index = FAISS.load_local(
    "/content/drive/MyDrive/farmlingua/vectorstore/live_rag_index",
    embedding_model,
    allow_dangerous_deserialization=True
)

static_faiss_index = FAISS.load_local(
    "/content/drive/MyDrive/farmlingua/vectorstore/faiss_index",
    embedding_model,
    allow_dangerous_deserialization=True
)


mistral_tokenizer = AutoTokenizer.from_pretrained("NousResearch/Nous-Hermes-2-Mistral-7B-DPO")
mistral_model = AutoModelForCausalLM.from_pretrained(
    "NousResearch/Nous-Hermes-2-Mistral-7B-DPO",
    torch_dtype=torch.float16,
    device_map="auto",
    trust_remote_code=True
)

mistral_pipe = pipeline(
    "text-generation",
    model=mistral_model,
    tokenizer=mistral_tokenizer,
    max_new_tokens=512,
    temperature=0.9,
    top_k=50,
    top_p=0.95,
    repetition_penalty=1.1,
    do_sample=True,
    return_full_text=False
)

expert_llm = HuggingFacePipeline(pipeline=mistral_pipe)

formatter_pipe = pipeline(
    "text2text-generation",
    model="google/flan-t5-large",
    max_length=256,
    temperature=0.3
)

formatter_llm = HuggingFacePipeline(pipeline=formatter_pipe)



def classify_intent(query):
    vect = vectorizer.transform([query])
    pred = intent_model.predict(vect)
    return label_encoder.inverse_transform(pred)[0]

def retrieve_context(intent, query):
    if intent in ["weather", "news", "update"]:
        docs = live_rag_index.similarity_search(query, k=5)
    else:
        docs = static_faiss_index.similarity_search(query, k=5)
    return "\n".join([doc.page_content for doc in docs])

user_agent = Agent(
    role="Farmer Interface Agent",
    goal="Receive query and trigger intent detection.",
    backstory="Frontline agent that interfaces with farmers across Nigeria.",
    verbose=True,
    allow_delegation=True
)

intent_agent = Agent(
    role="Intent Detection Agent",
    goal="Analyze farmer input and detect intent accurately using AI classifier.",
    backstory="Expert in classifying agricultural intent into soil, crop, weather, or news domains.",
    verbose=True,
    allow_delegation=True
)

retriever_agent = Agent(
    role="Knowledge Retrieval Agent",
    goal="Fetch relevant context only when needed for reasoning.",
    backstory="Knows how to pull useful documents from vectorstore if the query requires grounding.",
    verbose=True,
    allow_delegation=True
)

expert_agent = Agent(
    role="Agricultural Expert Agent",
    goal="Answer farmer questions clearly using reasoning, hallucination, or retrieval when needed.",
    backstory=(
        "You are a highly intelligent AI agricultural expert. You have deep reasoning skills, can hallucinate answers intelligently, "
        "and only use retrieved context if you are uncertain. Be smart, helpful, and explanatory."
    ),
    llm=expert_llm,
    verbose=True,
    allow_delegation=True
)

formatter_agent = Agent(
    role="Response Formatter Agent",
    goal="Polish and simplify the final answer into a clear, friendly explanation for farmers.",
    backstory="Specialist in structuring AI responses into simple, readable, and localized English for Nigerian farmers.",
    llm=formatter_llm,
    verbose=True
)


def create_crew(query):
    intent = classify_intent(query)
    context = retrieve_context(intent, query)

    expert_task = Task(
        description=f"""You're an agricultural AI advisor. The user asked: '{query}'.
Only use the context below **if necessary**:
========
{context}
========
Give a detailed, intelligent answer that may include reasoning, hallucination, and helpful explanation. Avoid repetition.
""",
        expected_output="A clear, helpful, and farmer-friendly answer to the user's query.",
        agent=expert_agent
    )

    format_task = Task(
        description="Take the previous response and reformat it into a simple, clean version suitable for a rural farmer.",
        expected_output="A clean, simplified explanation suitable for a rural farmer.",
        agent=formatter_agent
    )

    crew = Crew(
        agents=[user_agent, intent_agent, retriever_agent, expert_agent, formatter_agent],
        tasks=[expert_task, format_task],
        verbose=True
    )

    return crew


def run_pipeline(query: str):
    crew = create_crew(query)
    result = crew.kickoff()
    return result


response = run_pipeline("What is the best fertilizer for maize during rainy season?")
print(response)
```

```python
# app/agents/crew_pipeline.py


import os
import logging
import traceback
import math
from typing import List, Dict, Optional, Tuple

import joblib
import requests
import torch
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM
from langchain.vectorstores import FAISS
from langchain.embeddings import SentenceTransformerEmbeddings
from langchain.docstore.document import Document


CLASSIFIER_PATH = os.getenv(
    "INTENT_CLASSIFIER_PATH",
    "/content/drive/MyDrive/farmlingua_backend/app/models/intent_classifier_v2.joblib",
)
STATIC_VS_PATH = os.getenv(
    "STATIC_VS_PATH",
    "/content/drive/MyDrive/farmlingua_backend/app/vectorstore/faiss_index",
)
LIVE_VS_PATH = os.getenv(
    "LIVE_VS_PATH",
    "/content/drive/MyDrive/farmlingua_backend/app/vectorstore/live_rag_index",
)
EMBEDDING_MODEL = os.getenv(
    "EMBEDDING_MODEL", "sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2"
)
WEATHERAPI_KEY = os.getenv("WEATHERAPI_KEY", "d72aad0c21c74cf897b105451250408")

CLASSIFIER_CONFIDENCE_THRESHOLD = float(
    os.getenv("CLASSIFIER_CONFIDENCE_THRESHOLD", "0.6")
)

EXPERT_MODEL_NAME = os.getenv(
    "EXPERT_MODEL_NAME", "NousResearch/Nous-Hermes-2-Mistral-7B-DPO"
)
FORMATTER_MODEL_NAME = os.getenv("FORMATTER_MODEL_NAME", "google/flan-t5-large")


RAG_K = int(os.getenv("RAG_K", "4"))


logging.basicConfig(
    level=logging.INFO, format="%(asctime)s [%(levelname)s] %(name)s: %(message)s"
)
logger = logging.getLogger("crew_pipeline")


def safe_load_joblib(path: str):
    if not os.path.exists(path):
        raise FileNotFoundError(f"joblib file not found: {path}")
    return joblib.load(path)


def try_requests_get(url: str, params: dict = None, timeout: int = 10, retries: int = 2):
    for attempt in range(retries + 1):
        try:
            r = requests.get(url, params=params, timeout=timeout)
            r.raise_for_status()
            return r
        except Exception as e:
            logger.debug(f"requests.get attempt {attempt} failed for {url}: {e}")
            if attempt == retries:
                raise
    raise RuntimeError("unreachable")



class FarmlinguaCrew:
    def __init__(self):
        self.device = 0 if torch.cuda.is_available() else -1
        logger.info(f"Device -> {'cuda:0' if self.device == 0 else 'cpu'}")


        try:
            bundle = safe_load_joblib(CLASSIFIER_PATH)

            self.intent_model = bundle["model"]
            self.vectorizer = bundle["vectorizer"]
            self.label_encoder = bundle.get("label_encoder") or bundle.get("le")
            logger.info("Intent classifier loaded.")
        except Exception as e:
            logger.exception("Failed to load intent classifier.")
            raise


        try:
            self.embedder = SentenceTransformerEmbeddings(model_name=EMBEDDING_MODEL)

            if os.path.exists(STATIC_VS_PATH):
                self.static_vs = FAISS.load_local(
                    STATIC_VS_PATH,
                    self.embedder,
                    allow_dangerous_deserialization=True,
                )
                logger.info("Static FAISS vectorstore loaded.")
            else:
                self.static_vs = None
                logger.warning("Static vectorstore not found.")

            if os.path.exists(LIVE_VS_PATH):
                self.live_vs = FAISS.load_local(
                    LIVE_VS_PATH,
                    self.embedder,
                    allow_dangerous_deserialization=True,
                )
                logger.info("Live RAG FAISS vectorstore loaded.")
            else:
                self.live_vs = None
                logger.warning("Live vectorstore not found.")
        except Exception as e:
            logger.exception("Failed to load FAISS vectorstores.")
            raise


        try:
            logger.info(f"Loading expert LLM: {EXPERT_MODEL_NAME} (this may take a while)")
            expert_tok = AutoTokenizer.from_pretrained(EXPERT_MODEL_NAME, trust_remote_code=True)
            expert_model = AutoModelForCausalLM.from_pretrained(
                EXPERT_MODEL_NAME,
                torch_dtype=torch.float16,
                device_map="auto",
                trust_remote_code=True,
            )

            self.expert_pipe = pipeline(
                "text-generation",
                model=expert_model,
                tokenizer=expert_tok,
                trust_remote_code=True,

            )
            logger.info("Expert LLM ready.")
        except Exception as e:
            logger.exception("Failed to load expert LLM.")
            raise


        try:
            logger.info(f"Loading formatter LLM: {FORMATTER_MODEL_NAME}")

            formatter_device = 0 if (torch.cuda.is_available() and "cuda" in torch.device("cuda:0").type) else -1
            self.formatter_pipe = pipeline(
                "text2text-generation",
                model=FORMATTER_MODEL_NAME,
                tokenizer=FORMATTER_MODEL_NAME,
                device=formatter_device,
            )
            logger.info("Formatter LLM ready.")
        except Exception as e:
            logger.exception("Failed to load formatter LLM.")

            self.formatter_pipe = None


    def classify_intent(self, text: str) -> Tuple[str, float]:
        try:
            X = self.vectorizer.transform([text])

            if hasattr(self.intent_model, "predict_proba"):
                probs = self.intent_model.predict_proba(X)[0]
                idx = probs.argmax()
                conf = float(probs[idx])
            else:
                pred = self.intent_model.predict(X)[0]
                idx = pred
                conf = 1.0
            label = self.label_encoder.inverse_transform([idx])[0]
            logger.debug(f"Intent classified -> {label} (conf {conf:.3f})")
            return label, conf
        except Exception:
            logger.exception("Intent classification failed.")
            return "general_agriculture", 0.0


    def retrieve_docs(self, query: str, intent: str, k: int = RAG_K) -> List[Document]:
        docs = []
        try:
            if intent in ("weather_forecast", "weather"):
                if self.live_vs:
                    docs = self.live_vs.similarity_search(query, k=k)
            elif intent in ("news", "update", "agri_news"):
                if self.live_vs:
                    docs = self.live_vs.similarity_search(query, k=k)
                elif self.static_vs:
                    docs = self.static_vs.similarity_search(query, k=k)
            else:

                if self.static_vs:
                    docs = self.static_vs.similarity_search(query, k=k)
                elif self.live_vs:
                    docs = self.live_vs.similarity_search(query, k=k)
            logger.debug(f"Retrieved {len(docs)} docs for intent {intent}")
        except AssertionError as ae:

            logger.exception(
                "FAISS assertion — likely embedding dim mismatch. Check that the same embedding model "
                "was used to create & load the index."
            )
        except Exception:
            logger.exception("Failed to retrieve docs from vectorstore.")
        return docs


    def get_weather_for_location(self, location: str) -> Optional[str]:
        if not WEATHERAPI_KEY:
            logger.warning("No WEATHERAPI_KEY configured; cannot fetch live weather.")
            return None
        try:
            url = "http://api.weatherapi.com/v1/current.json"
            params = {"key": WEATHERAPI_KEY, "q": location}
            r = try_requests_get(url, params=params, timeout=8, retries=2)
            data = r.json()
            if "current" in data:
                c = data["current"]
                condition = c["condition"]["text"]
                temp = c["temp_c"]
                hum = c["humidity"]
                return f"Weather in {location}: {condition}, Temp {temp}°C, Humidity {hum}%"
            return None
        except Exception:
            logger.exception("WeatherAPI failure")
            return None


    def generate_answer(self, query: str, intent: str, context_docs: List[Document]) -> Dict:

        context_str = ""
        sources = []
        if context_docs:
            snippets = []
            for d in context_docs:
                snippets.append(d.page_content.strip())
                src = d.metadata.get("source") or d.metadata.get("location") or "unknown"
                sources.append(str(src))
            context_str = "\n\n".join(snippets[:6])


        if intent in ("weather_forecast", "weather"):

            #
            tokens = [t.strip(",.?") for t in query.split()]
            candidate = None
            for t in reversed(tokens):
                if t[0].isupper() or t.lower() in ["lagos", "kaduna", "rivers", "edo"]:
                    candidate = t
                    break

            location = candidate or "Lagos"
            weather_text = self.get_weather_for_location(location)
            if weather_text:
                prompt = f"Question: {query}\n\nWeather data: {weather_text}\n\nAnswer:"
            else:

                prompt = f"Question: {query}\n\nContext:\n{context_str}\n\nAnswer:"
        else:

            prompt = f"Question: {query}\n\n"
            if context_str:
                prompt += f"Context:\n{context_str}\n\n"
            prompt += "Answer (be concise, actionable, and farmer-friendly):"


        try:

            gen_kwargs = dict(
                do_sample=True,
                top_p=0.95,
                top_k=40,
                temperature=0.7,
                max_new_tokens=256,
                repetition_penalty=1.05,
            )

            raw = self.expert_pipe(prompt, **gen_kwargs)

            gen_text = ""
            if isinstance(raw, list) and len(raw) > 0:
                first = raw[0]
                gen_text = first.get("generated_text") or first.get("text") or str(first)
            else:
                gen_text = str(raw)
        except Exception:
            logger.exception("Expert LLM generation failed; returning fallback message.")
            gen_text = "Sorry — I couldn't generate an answer at the moment."


        formatted = None
        try:
            if self.formatter_pipe:
                format_prompt = (
                    "Rewrite the following answer into a short, clear, and friendly explanation for a "
                    "smallholder farmer in Nigeria. Keep it actionable and in plain English:\n\n"
                    + gen_text
                )
                fmt_raw = self.formatter_pipe(format_prompt, max_length=200)
                if isinstance(fmt_raw, list) and fmt_raw:
                    formatted = fmt_raw[0].get("generated_text") or fmt_raw[0].get("text")
                else:
                    formatted = str(fmt_raw)
        except Exception:
            logger.exception("Formatter LLM failed; falling back to raw expert text.")
            formatted = None

        result = {
            "query": query,
            "intent": intent,
            "sources": list(dict.fromkeys(sources)),
            "raw_answer": gen_text.strip(),
            "answer": (formatted or gen_text).strip(),
        }
        return result


    def handle(self, query: str) -> Dict:
        try:
            if not query or not isinstance(query, str) or not query.strip():
                return {"error": "invalid query", "answer": ""}


            intent, conf = self.classify_intent(query)


            need_rag = False
            if intent in ("weather_forecast", "weather", "news", "update", "agri_news"):
                need_rag = True
            elif conf < CLASSIFIER_CONFIDENCE_THRESHOLD:
                need_rag = True

            docs = []
            if need_rag:
                docs = self.retrieve_docs(query, intent, k=RAG_K)


            if intent in ("weather_forecast", "weather") and not docs:

                pass


            result = self.generate_answer(query, intent, docs)

            result["classifier_confidence"] = conf
            return result
        except Exception:
            logger.exception("Top-level handle failed.")
            return {"error": "internal_error", "trace": traceback.format_exc()}

if __name__ == "__main__":

    crew = FarmlinguaCrew()
    tests = [
        "Will it rain in Lagos this weekend?",
        "How can I improve soil for cassava?",
        "My tomatoes have black spots — what disease could this be?",
    ]
    for q in tests:
        out = crew.handle(q)
        print("---")
        print("Query:", q)
        print("Intent:", out.get("intent"), "conf:", out.get("classifier_confidence"))
        print("Sources:", out.get("sources"))
        print("Answer:\n", out.get("answer"))
```

```python
# farmlingua/app/agents/crew_pipeline.py
import os
import sys
import requests
import joblib
import faiss
import numpy as np
from transformers import pipeline
from sentence_transformers import SentenceTransformer
from app.utils import config
from datetime import datetime, timedelta
from collections import defaultdict


hf_cache = "/models/huggingface"
os.environ["HF_HOME"] = hf_cache
os.environ["TRANSFORMERS_CACHE"] = hf_cache
os.environ["HUGGINGFACE_HUB_CACHE"] = hf_cache
os.makedirs(hf_cache, exist_ok=True)

BASE_DIR = os.path.dirname(os.path.dirname(os.path.abspath(__file__)))
if BASE_DIR not in sys.path:
    sys.path.insert(0, BASE_DIR)

DEVICE = 0 if os.environ.get("CUDA_VISIBLE_DEVICES") else -1

try:
    classifier = joblib.load(config.CLASSIFIER_PATH)
except Exception:
    classifier = None


print(f"Loading expert model ({config.EXPERT_MODEL_NAME})...")
expert_pipeline = pipeline(
    "text-generation",
    model=config.EXPERT_MODEL_NAME,
    device=DEVICE,
    max_new_tokens=500,
    temperature=0.3,
    repetition_penalty=1.1
)

print(f"Loading formatter model ({config.FORMATTER_MODEL_NAME})...")
formatter_pipeline = pipeline(
    "text2text-generation",
    model=config.FORMATTER_MODEL_NAME,
    device=DEVICE
)


embedder = SentenceTransformer(config.EMBEDDING_MODEL)

session_memory = defaultdict(dict)
SESSION_TIMEOUT = timedelta(hours=1)
MAX_TURNS = 5

def get_session(session_id: str):
    """Retrieve session memory or reset if expired."""
    session = session_memory.get(session_id, {})
    if not session or session.get('expires_at') < datetime.now():
        session_memory[session_id] = {
            'messages': [],
            'expires_at': datetime.now() + SESSION_TIMEOUT
        }
    return session_memory[session_id]

def add_to_session(session_id: str, role: str, content: str):
    """Store conversation turn in short-term memory."""
    session = get_session(session_id)
    session['messages'].append({'role': role, 'content': content})

    if len(session['messages']) > MAX_TURNS:
        session['messages'] = session['messages'][-MAX_TURNS:]
    session['expires_at'] = datetime.now() + SESSION_TIMEOUT

def get_short_term_context(session_id: str):
    """Return last conversation context for follow-ups."""
    session = get_session(session_id)
    history = session.get('messages', [])
    return "\n".join(
        f"{m['role'].upper()}: {m['content']}" for m in history
    )



def retrieve_docs(query, vs_path):
    if not vs_path or not os.path.exists(vs_path):
        return None

    if os.path.isdir(vs_path):
        try:
            from langchain.vectorstores import FAISS as LCFAISS
            from langchain.embeddings import SentenceTransformerEmbeddings
            embed_model = SentenceTransformerEmbeddings(model_name=config.EMBEDDING_MODEL)
            vs = LCFAISS.load_local(str(vs_path), embed_model, allow_dangerous_deserialization=True)
            docs = vs.similarity_search(query, k=3)
            return "\n\n".join(d.page_content for d in docs) if docs else None
        except Exception:
            return None

    try:
        index = faiss.read_index(str(vs_path))
    except Exception:
        return None

    query_vec = np.array([embedder.encode(query)], dtype=np.float32)
    D, I = index.search(query_vec, k=3)
    if D[0][0] == 0:
        return None

    meta_path = str(vs_path) + "_meta.npy"
    if os.path.exists(meta_path):
        metadata = np.load(meta_path, allow_pickle=True).item()
        docs = [metadata.get(str(idx), "") for idx in I[0] if str(idx) in metadata]
        docs = [doc for doc in docs if doc]
        return "\n\n".join(docs) if docs else None
    return None



def get_weather(state_name):
    url = "http://api.weatherapi.com/v1/current.json"
    params = {
        "key": config.WEATHER_API_KEY,
        "q": f"{state_name}, Nigeria",
        "aqi": "no"
    }
    r = requests.get(url, params=params)
    if r.status_code != 200:
        return f"Unable to retrieve weather for {state_name}."
    data = r.json()
    return (
        f"Weather in {state_name}:\n"
        f"- Condition: {data['current']['condition']['text']}\n"
        f"- Temperature: {data['current']['temp_c']}°C\n"
        f"- Humidity: {data['current']['humidity']}%\n"
        f"- Wind: {data['current']['wind_kph']} kph"
    )


def check_self_identity(query: str):
    q_lower = query.lower()
    if any(word in q_lower for word in ["who built you", "who created you", "your creator", "who made you"]):
        return (
            "I was designed and built by an AI engineer from Remostart, "
            "a top AI company focused on using artificial intelligence to support farmers."
        )
    if any(word in q_lower for word in ["your purpose", "your role", "what are you", "what do you do"]):
        return (
            "I am an agricultural assistant built to make farming easier. "
            "I guide Nigerian farmers by answering questions, providing weather updates, "
            "and sharing farming best practices."
        )
    return None


def detect_intent(query):
    q_lower = query.lower()
    if any(word in q_lower for word in ["weather", "temperature", "rain", "forecast"]):
        for state in config.STATES:
            if state.lower() in q_lower:
                return "weather", state
        return "weather", None
    if any(word in q_lower for word in ["latest", "update", "breaking", "news", "current", "predict"]):
        return "live_update", None
    if hasattr(classifier, "predict") and hasattr(classifier, "predict_proba"):
        predicted_intent = classifier.predict([query])[0]
        confidence = max(classifier.predict_proba([query])[0])
        if confidence < config.CLASSIFIER_CONFIDENCE_THRESHOLD:
            return "low_confidence", None
        return predicted_intent, None
    return "normal", None



def run_pipeline(user_query: str, session_id: str = "default"):

    self_response = check_self_identity(user_query)
    if self_response:
        return self_response


    intent, extra = detect_intent(user_query)


    if intent == "weather" and extra:
        weather_text = get_weather(extra)
        return formatter_pipeline(
            (
                "Rewrite this weather update in simple farmer-friendly language. "
                "If it has multiple conditions or data points, present them as a clean list with 1., 2., 3.\n\n"
                f"{weather_text}"
            ),
            max_length=256,
            do_sample=False
        )[0]["generated_text"]


    if intent == "live_update":
        context = retrieve_docs(user_query, config.LIVE_VS_PATH)
        if context:
            user_query += f"\n\nLatest agricultural updates:\n{context}"

    if intent == "low_confidence":
        context = retrieve_docs(user_query, config.STATIC_VS_PATH)
        if context:
            user_query += f"\n\nReference information:\n{context}"


    history_context = get_short_term_context(session_id)
    if history_context:
        user_query = f"Conversation so far:\n{history_context}\n\nCurrent question: {user_query}"


    expert_response = expert_pipeline(
        f"Provide a detailed agricultural answer for Nigerian farmers:\n\n{user_query}",
        max_new_tokens=500,
        temperature=0.3
    )[0]['generated_text']


    formatted_response = formatter_pipeline(
        f"""
Rewrite the following expert answer so it is:
- Clear, simple, and farmer-friendly (avoid technical jargon).
- Structured with short sentences and easy explanations, ChatGPT-style.
- If listing steps/solutions, use numbered points (1., 2., 3.) with each on a new line.
- Do NOT include irrelevant details (like chemical elements).
- Make it sound like practical advice for Nigerian farmers.

Answer to format:
{expert_response}
""",
        max_length=512,
        do_sample=False
    )[0]['generated_text']


    add_to_session(session_id, "user", user_query)
    add_to_session(session_id, "assistant", formatted_response)

    return formatted_response
```

```python
# farmlingua/app/agents/crew_pipeline.py
import os
import sys
import requests
import joblib
import faiss
import numpy as np
from transformers import pipeline
from sentence_transformers import SentenceTransformer
from app.utils import config


hf_cache = "/models/huggingface"
os.environ["HF_HOME"] = hf_cache
os.environ["TRANSFORMERS_CACHE"] = hf_cache
os.environ["HUGGINGFACE_HUB_CACHE"] = hf_cache
os.makedirs(hf_cache, exist_ok=True)

BASE_DIR = os.path.dirname(os.path.dirname(os.path.abspath(__file__)))
if BASE_DIR not in sys.path:
    sys.path.insert(0, BASE_DIR)

DEVICE = 0 if os.environ.get("CUDA_VISIBLE_DEVICES") else -1

try:
    classifier = joblib.load(config.CLASSIFIER_PATH)
except Exception:
    classifier = None


print(f"Loading expert model ({config.EXPERT_MODEL_NAME})...")
expert_pipeline = pipeline(
    "text-generation",
    model=config.EXPERT_MODEL_NAME,
    device=DEVICE,
    max_new_tokens=500,
    temperature=0.3,
    repetition_penalty=1.1
)

print(f"Loading formatter model ({config.FORMATTER_MODEL_NAME})...")
formatter_pipeline = pipeline(
    "text2text-generation",
    model=config.FORMATTER_MODEL_NAME,
    device=DEVICE
)


embedder = SentenceTransformer(config.EMBEDDING_MODEL)

def retrieve_docs(query, vs_path):
    if not vs_path or not os.path.exists(vs_path):
        return None

    if os.path.isdir(vs_path):
        try:
            from langchain.vectorstores import FAISS as LCFAISS
            from langchain.embeddings import SentenceTransformerEmbeddings
            embed_model = SentenceTransformerEmbeddings(model_name=config.EMBEDDING_MODEL)
            vs = LCFAISS.load_local(str(vs_path), embed_model, allow_dangerous_deserialization=True)
            docs = vs.similarity_search(query, k=3)
            return "\n\n".join(d.page_content for d in docs) if docs else None
        except Exception:
            return None

    try:
        index = faiss.read_index(str(vs_path))
    except Exception:
        return None

    query_vec = np.array([embedder.encode(query)], dtype=np.float32)
    D, I = index.search(query_vec, k=3)
    if D[0][0] == 0:
        return None

    meta_path = str(vs_path) + "_meta.npy"
    if os.path.exists(meta_path):
        metadata = np.load(meta_path, allow_pickle=True).item()
        docs = [metadata.get(str(idx), "") for idx in I[0] if str(idx) in metadata]
        docs = [doc for doc in docs if doc]
        return "\n\n".join(docs) if docs else None
    return None


def get_weather(state_name):
    url = "http://api.weatherapi.com/v1/current.json"
    params = {
        "key": config.WEATHER_API_KEY,
        "q": f"{state_name}, Nigeria",
        "aqi": "no"
    }
    r = requests.get(url, params=params)
    if r.status_code != 200:
        return f"Unable to retrieve weather for {state_name}."
    data = r.json()
    return (
        f"Weather in {state_name}:\n"
        f"- Condition: {data['current']['condition']['text']}\n"
        f"- Temperature: {data['current']['temp_c']}°C\n"
        f"- Humidity: {data['current']['humidity']}%\n"
        f"- Wind: {data['current']['wind_kph']} kph"
    )



def check_self_identity(query: str):
    q_lower = query.lower()
    if any(word in q_lower for word in ["who built you", "who created you", "your creator", "who made you"]):
        return (
            "I was designed and built by an AI engineer from Remostart a top AI company. "
            "The goal is to use artificial intelligence to support farmers with practical agricultural advice."
        )
    if any(word in q_lower for word in ["what do you do", "your purpose", "your role", "what are you"]):
        return (
            "I am an agricultural assistant built to make farming easier. "
            "I guide Nigerian farmers by answering their questions, providing weather updates, and sharing farming best practices."
        )
    return None


def detect_intent(query):
    q_lower = query.lower()
    if any(word in q_lower for word in ["weather", "temperature", "rain", "forecast"]):
        for state in config.STATES:
            if state.lower() in q_lower:
                return "weather", state
        return "weather", None
    if any(word in q_lower for word in ["latest", "update", "breaking", "news", "current", "predict"]):
        return "live_update", None
    if hasattr(classifier, "predict") and hasattr(classifier, "predict_proba"):
        predicted_intent = classifier.predict([query])[0]
        confidence = max(classifier.predict_proba([query])[0])
        if confidence < config.CLASSIFIER_CONFIDENCE_THRESHOLD:
            return "low_confidence", None
        return predicted_intent, None
    return "normal", None



def run_pipeline(user_query: str):

    self_response = check_self_identity(user_query)
    if self_response:
        return self_response

    intent, extra = detect_intent(user_query)


    if intent == "weather" and extra:
        weather_text = get_weather(extra)
        return formatter_pipeline(
            (
                "Rewrite this weather update in simple farmer-friendly language. "
                "If it has multiple conditions or data points, present them as a clean list with 1., 2., 3. "
                "Do not use bullet points for normal sentences.\n\n"
                f"{weather_text}"
            ),
            max_length=256,
            do_sample=False
        )[0]["generated_text"]


    if intent == "live_update":
        context = retrieve_docs(user_query, config.LIVE_VS_PATH)
        if context:
            user_query += f"\n\nLatest agricultural updates:\n{context}"


    if intent == "low_confidence":
        context = retrieve_docs(user_query, config.STATIC_VS_PATH)
        if context:
            user_query += f"\n\nReference information:\n{context}"


    expert_response = expert_pipeline(
        f"Provide a detailed agricultural answer for Nigerian farmers:\n\n{user_query}",
        max_new_tokens=500,
        temperature=0.3
    )[0]['generated_text']


    formatted_response = formatter_pipeline(
        f"""
Rewrite the following expert answer so it is:
- Clear, simple, and farmer-friendly (avoid technical jargon) for local nigeria farmers.
- Structured with short sentences and easy explanations, don't join everything together structure it like Chatgpt response style.
- If listing solutions or steps, use numbered points (1., 2., 3.) and let it be line by line different lines for each point not joined together in one line.
- Do NOT dump irrelevant details (like chemical elements not useful for farming).
- Make it sound like practical advice for Nigerian farmers, make sure all sentences are complete and not incomplete.

Answer to format:
{expert_response}
""",
        max_length=512,
        do_sample=False
    )[0]['generated_text']

    return formatted_response
```

```python

from datasets import load_dataset
from transformers import AutoTokenizer, AutoModelForCausalLM, TrainingArguments, Trainer
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training
import torch


ds1 = load_dataset("YuvrajSingh9886/Agriculture-Plan-Diseases-QA-Pairs-Dataset")


def format_row(example):
    return {
        "instruction": example["QUESTION.question"],
        "input": "",
        "output": example["ANSWER"]
    }


dataset = ds1["train"].map(format_row)
dataset = dataset.shuffle(seed=42).select(range(1000))


model_id = "persadian/CropSeek-LLM"
tokenizer = AutoTokenizer.from_pretrained(model_id, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",
    load_in_4bit=True,
    torch_dtype=torch.float16,
    trust_remote_code=True
)


model = prepare_model_for_kbit_training(model)


lora_config = LoraConfig(
    r=8,
    lora_alpha=32,
    target_modules=["q_proj", "v_proj"],
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM"
)
model = get_peft_model(model, lora_config)


def tokenize(batch):
    prompts = [
        f"### Instruction:\n{inst}\n\n### Response:\n{out}"
        for inst, out in zip(batch["instruction"], batch["output"])
    ]
    encodings = tokenizer(
        prompts,
        truncation=True,
        padding="max_length",
        max_length=512,
        return_tensors="pt"
    )
    result = {k: v.tolist() for k, v in encodings.items()}
    result["labels"] = result["input_ids"].copy()
    return result


tokenized = dataset.map(tokenize, batched=True, remove_columns=dataset.column_names)


training_args = TrainingArguments(
    output_dir="/content/farmlingua_cropseek_lora",
    per_device_train_batch_size=2,
    gradient_accumulation_steps=4,
    num_train_epochs=3,
    learning_rate=2e-4,
    fp16=True,
    logging_steps=20,
    save_strategy="epoch",
    report_to="none"
)


trainer = Trainer(
    model=model,
    tokenizer=tokenizer,
    args=training_args,
    train_dataset=tokenized
)

trainer.train()


model.save_pretrained("/content/farmlingua_cropseek_lora")
tokenizer.save_pretrained("/content/farmlingua_cropseek_lora")


!zip -r /content/farmlingua_cropseek_lora.zip /content/farmlingua_cropseek_lora
```

```python
# farmlingua/app/agents/crew_pipeline.pyversion3multilingua
import os
import sys
import requests
import joblib
import faiss
import numpy as np
import torch
import fasttext
from huggingface_hub import hf_hub_download
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
from sentence_transformers import SentenceTransformer
from app.utils import config
import re


hf_cache = "/models/huggingface"
os.environ["HF_HOME"] = hf_cache
os.environ["TRANSFORMERS_CACHE"] = hf_cache
os.environ["HUGGINGFACE_HUB_CACHE"] = hf_cache
os.makedirs(hf_cache, exist_ok=True)

BASE_DIR = os.path.dirname(os.path.dirname(os.path.abspath(__file__)))
if BASE_DIR not in sys.path:
    sys.path.insert(0, BASE_DIR)

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"


try:
    classifier = joblib.load(config.CLASSIFIER_PATH)
except Exception:
    classifier = None


print(f"Loading Qwen expert model ({config.EXPERT_MODEL_NAME})...")
tokenizer = AutoTokenizer.from_pretrained(config.EXPERT_MODEL_NAME)
model = AutoModelForCausalLM.from_pretrained(
    config.EXPERT_MODEL_NAME,
    torch_dtype="auto",
    device_map="auto"
)

embedder = SentenceTransformer(config.EMBEDDING_MODEL)


print(f"Loading FastText language identifier ({config.LANG_ID_MODEL_REPO})...")
lang_model_path = hf_hub_download(
    repo_id=config.LANG_ID_MODEL_REPO,
    filename=config.LANG_ID_MODEL_FILE
)
lang_identifier = fasttext.load_model(lang_model_path)

def detect_language(text, top_k=1):
    """Detect language with FastText, ensuring no newlines."""
    clean_text = text.replace("\n", " ").strip()
    labels, probs = lang_identifier.predict(clean_text, k=top_k)
    return [(l.replace("__label__", ""), float(p)) for l, p in zip(labels, probs)]


print(f"Loading NLLB translation model ({config.TRANSLATION_MODEL_NAME})...")
translation_pipeline = pipeline(
    "translation",
    model=config.TRANSLATION_MODEL_NAME,
    device=0 if DEVICE == "cuda" else -1,
    max_new_tokens=512  # allow longer translations
)

SUPPORTED_LANGS = {
    "eng_Latn": "English",
    "ibo_Latn": "Igbo",
    "yor_Latn": "Yoruba",
    "hau_Latn": "Hausa",
    "swh_Latn": "Swahili",
    "amh_Ethi": "Amharic"
}


# Translation Helper with Chunking
def chunk_text(text, max_len=400):
    """Split text into chunks without cutting sentences."""
    sentences = re.split(r'(?<=[.!?]) +', text)
    chunks, current = [], ""

    for sent in sentences:
        if len(current) + len(sent) < max_len:
            current += " " + sent
        else:
            chunks.append(current.strip())
            current = sent
    if current:
        chunks.append(current.strip())
    return chunks


def translate_text(text, src_lang, tgt_lang):
    """Translate with chunking and stitch results together."""
    if not text.strip():
        return text
    chunks = chunk_text(text)
    results = []
    for chunk in chunks:
        out = translation_pipeline(chunk, src_lang=src_lang, tgt_lang=tgt_lang)
        results.append(out[0]['translation_text'])
    return " ".join(results)


# RAG Retrieval
def retrieve_docs(query, vs_path):
    if not vs_path or not os.path.exists(vs_path):
        return None
    try:
        index = faiss.read_index(vs_path)
    except Exception:
        return None
    query_vec = np.array([embedder.encode(query)], dtype=np.float32)
    D, I = index.search(query_vec, k=3)
    if D[0][0] == 0:
        return None
    meta_path = vs_path + "_meta.npy"
    if os.path.exists(meta_path):
        metadata = np.load(meta_path, allow_pickle=True).item()
        docs = [metadata.get(str(idx), "") for idx in I[0] if str(idx) in metadata]
        docs = [doc for doc in docs if doc]
        return "\n\n".join(docs) if docs else None
    return None


# Weather
def get_weather(state_name):
    url = "http://api.weatherapi.com/v1/current.json"
    params = {
        "key": config.WEATHER_API_KEY,
        "q": f"{state_name}, Nigeria",
        "aqi": "no"
    }
    r = requests.get(url, params=params)
    if r.status_code != 200:
        return f"Unable to retrieve weather for {state_name}."
    data = r.json()
    return (
        f"Weather in {state_name}:\n"
        f"- Condition: {data['current']['condition']['text']}\n"
        f"- Temperature: {data['current']['temp_c']}°C\n"
        f"- Humidity: {data['current']['humidity']}%\n"
        f"- Wind: {data['current']['wind_kph']} kph"
    )


# Intent Detection
def detect_intent(query):
    q_lower = query.lower()

    if any(word in q_lower for word in ["weather condition", "forecast"]):
        for state in config.STATES:
            if state.lower() in q_lower:
                return "weather", state
        return "weather", None

    if any(word in q_lower for word in ["update", "breaking", "news", "current"]):
        return "live_update", None

    if hasattr(classifier, "predict") and hasattr(classifier, "predict_proba"):
        predicted_intent = classifier.predict([query])[0]
        confidence = max(classifier.predict_proba([query])[0])
        if confidence < config.CLASSIFIER_CONFIDENCE_THRESHOLD:
            return "low_confidence", None
        return predicted_intent, None

    return "normal", None


#  Qwen Runner
def run_qwen(messages, max_new_tokens=1000):
    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )
    inputs = tokenizer([text], return_tensors="pt").to(model.device)

    generated_ids = model.generate(
        **inputs,
        max_new_tokens=max_new_tokens,
        temperature=0.4,
        repetition_penalty=1.1
    )
    output_ids = generated_ids[0][len(inputs.input_ids[0]):].tolist()
    return tokenizer.decode(output_ids, skip_special_tokens=True).strip()


# Main Pipeline
def run_pipeline(user_query: str):
    lang, prob = detect_language(user_query, top_k=1)[0]
    if lang not in SUPPORTED_LANGS:
        lang = "eng_Latn"

    # Translate to English if needed
    translated_query = user_query
    if lang != "eng_Latn":
        translated_query = translate_text(user_query, src_lang=lang, tgt_lang="eng_Latn")

    # Detect intent
    intent, extra = detect_intent(translated_query)

    if intent == "weather" and extra:
        weather_text = get_weather(extra)
        messages = [
            {"role": "system", "content": "You are FarmLingua, an AI assistant for Nigerian farmers."},
            {"role": "user", "content": f"Rewrite this weather update simply:\n{weather_text}"}
        ]
        english_answer = run_qwen(messages, max_new_tokens=256)
    else:
        if intent == "live_update":
            context = retrieve_docs(translated_query, config.LIVE_VS_PATH)
            if context:
                translated_query += f"\n\nLatest agricultural updates:\n{context}"

        if intent == "low_confidence":
            context = retrieve_docs(translated_query, config.STATIC_VS_PATH)
            if context:
                translated_query += f"\n\nReference information:\n{context}"

        messages = [
            {"role": "system", "content": (
                "You are FarmLingua, an AI assistant for Nigerian farmers. "
                "Answer directly without repeating the question. "
                "Use short, clear farmer-friendly English. "
                "Avoid scientific jargon, focus on practical farming advice."
            )},
            {"role": "user", "content": translated_query}
        ]
        english_answer = run_qwen(messages, max_new_tokens=700)

    # Translate back to original language
    if lang != "eng_Latn":
        final_answer = translate_text(english_answer, src_lang="eng_Latn", tgt_lang=lang)
    else:
        final_answer = english_answer

    return {
        "detected_language": SUPPORTED_LANGS.get(lang, "Unknown"),
        "answer": final_answer
    }
```

```python
# app/utils/memory.py

from cachetools import TTLCache
from threading import Lock

# TTLCache(maxsize, ttl_seconds)
# - maxsize: maximum number of sessions stored
# - ttl: time-to-live in seconds (1 hour = 3600s)
memory_cache = TTLCache(maxsize=1000, ttl=3600)
lock = Lock()


class MemoryStore:
    """In-memory conversational history with 1-hour expiry."""

    def get_history(self, session_id: str):
        """Retrieve conversation history (list of messages)."""
        with lock:
            return memory_cache.get(session_id, []).copy()

    def save_history(self, session_id: str, history: list):
        """Save/overwrite conversation history."""
        with lock:
            memory_cache[session_id] = history.copy()

    def clear_history(self, session_id: str):
        """Manually clear a session."""
        with lock:
            if session_id in memory_cache:
                del memory_cache[session_id]

    def clear_all(self):
        """Clear all sessions (not usually needed in production)."""
        with lock:
            memory_cache.clear()


# Singleton store
memory_store = MemoryStore()
```

```python
# farmlingua/app/agents/crew_pipeline.py
import os
import sys
import re
import uuid
import requests
import joblib
import faiss
import numpy as np
import torch
import fasttext
from huggingface_hub import hf_hub_download
from transformers import AutoTokenizer, AutoModelForCausalLM, AutoModelForSeq2SeqLM, NllbTokenizer
from sentence_transformers import SentenceTransformer
from app.utils import config
from app.utils.memory import memory_store  # memory module
from typing import List


hf_cache = "/models/huggingface"
os.environ["HF_HOME"] = hf_cache
os.environ["TRANSFORMERS_CACHE"] = hf_cache
os.environ["HUGGINGFACE_HUB_CACHE"] = hf_cache
os.makedirs(hf_cache, exist_ok=True)

BASE_DIR = os.path.dirname(os.path.dirname(os.path.abspath(__file__)))
if BASE_DIR not in sys.path:
    sys.path.insert(0, BASE_DIR)

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"


try:
    classifier = joblib.load(config.CLASSIFIER_PATH)
except Exception:
    classifier = None


print(f"Loading expert model ({config.EXPERT_MODEL_NAME})...")
tokenizer = AutoTokenizer.from_pretrained(config.EXPERT_MODEL_NAME, use_fast=False)
model = AutoModelForCausalLM.from_pretrained(
    config.EXPERT_MODEL_NAME,
    torch_dtype="auto",
    device_map="auto"
)


embedder = SentenceTransformer(config.EMBEDDING_MODEL)

#   language detector
print(f"Loading FastText language identifier ({config.LANG_ID_MODEL_REPO})...")
lang_model_path = hf_hub_download(
    repo_id=config.LANG_ID_MODEL_REPO,
    filename=getattr(config, "LANG_ID_MODEL_FILE", "model.bin")
)
lang_identifier = fasttext.load_model(lang_model_path)

def detect_language(text: str, top_k: int = 1):
    if not text or not text.strip():
        return [("eng_Latn", 1.0)]
    clean_text = text.replace("\n", " ").strip()
    labels, probs = lang_identifier.predict(clean_text, k=top_k)
    return [(l.replace("__label__", ""), float(p)) for l, p in zip(labels, probs)]

#  Translation model
print(f"Loading translation model ({config.TRANSLATION_MODEL_NAME})...")

translation_tokenizer = NllbTokenizer.from_pretrained(
    config.TRANSLATION_MODEL_NAME,
    cache_dir=hf_cache
)

translation_model = AutoModelForSeq2SeqLM.from_pretrained(
    config.TRANSLATION_MODEL_NAME,
    torch_dtype=torch.float16 if DEVICE == "cuda" else torch.float32,
    cache_dir=hf_cache
)

if DEVICE == "cuda":
    translation_model = translation_model.to("cuda")
else:
    translation_model = translation_model.to("cpu")

print(f"Translation model loaded on {DEVICE}")

SUPPORTED_LANGS = {
    "eng_Latn": "English",
    "ibo_Latn": "Igbo",
    "yor_Latn": "Yoruba",
    "hau_Latn": "Hausa",
    "swh_Latn": "Swahili",
    "amh_Latn": "Amharic",
}

# Text chunking
_SENTENCE_SPLIT_RE = re.compile(r'(?<=[.!?])\s+')

def chunk_text(text: str, max_len: int = 400) -> List[str]:
    if not text:
        return []
    sentences = _SENTENCE_SPLIT_RE.split(text)
    chunks, current = [], ""
    for s in sentences:
        if not s:
            continue
        if len(current) + len(s) + 1 <= max_len:
            current = (current + " " + s).strip()
        else:
            if current:
                chunks.append(current.strip())
            current = s.strip()
    if current:
        chunks.append(current.strip())
    return chunks

def translate_text(text: str, src_lang: str, tgt_lang: str, max_chunk_len: int = 400) -> str:
    """Translate text using NLLB model"""
    if not text.strip():
        return text

    if src_lang == tgt_lang:
        return text

    chunks = chunk_text(text, max_len=max_chunk_len)
    translated_parts = []

    for chunk in chunks:

        translation_tokenizer.src_lang = src_lang

        # Tokenize
        inputs = translation_tokenizer(
            chunk,
            return_tensors="pt",
            padding=True,
            truncation=True,
            max_length=512
        ).to(translation_model.device)


        forced_bos_token_id = translation_tokenizer.convert_tokens_to_ids(tgt_lang)

        # Generate translation
        generated_tokens = translation_model.generate(
            **inputs,
            forced_bos_token_id=forced_bos_token_id,
            max_new_tokens=512,
            num_beams=5,
            early_stopping=True
        )

        # Decode
        translated_text = translation_tokenizer.batch_decode(
            generated_tokens,
            skip_special_tokens=True
        )[0]

        translated_parts.append(translated_text)

    return " ".join(translated_parts).strip()


#  RAG retrieval
def retrieve_docs(query: str, vs_path: str):
    if not vs_path or not os.path.exists(vs_path):
        return None
    try:
        index = faiss.read_index(str(vs_path))
    except Exception:
        return None
    query_vec = np.array([embedder.encode(query)], dtype=np.float32)
    D, I = index.search(query_vec, k=3)
    if D[0][0] == 0:
        return None
    meta_path = str(vs_path) + "_meta.npy"
    if os.path.exists(meta_path):
        metadata = np.load(meta_path, allow_pickle=True).item()
        docs = [metadata.get(str(idx), "") for idx in I[0] if str(idx) in metadata]
        docs = [d for d in docs if d]
        return "\n\n".join(docs) if docs else None
    return None


def get_weather(state_name: str) -> str:
    url = "http://api.weatherapi.com/v1/current.json"
    params = {"key": config.WEATHER_API_KEY, "q": f"{state_name}, Nigeria", "aqi": "no"}
    r = requests.get(url, params=params, timeout=10)
    if r.status_code != 200:
        return f"Unable to retrieve weather for {state_name}."
    data = r.json()
    return (
        f"Weather in {state_name}:\n"
        f"- Condition: {data['current']['condition']['text']}\n"
        f"- Temperature: {data['current']['temp_c']}°C\n"
        f"- Humidity: {data['current']['humidity']}%\n"
        f"- Wind: {data['current']['wind_kph']} kph"
    )


def detect_intent(query: str):
    q_lower = (query or "").lower()
    if any(word in q_lower for word in ["weather", "temperature", "rain", "forecast"]):
        for state in getattr(config, "STATES", []):
            if state.lower() in q_lower:
                return "weather", state
        return "weather", None

    if any(word in q_lower for word in ["latest", "update", "breaking", "news", "current", "predict"]):
        return "live_update", None

    if hasattr(classifier, "predict") and hasattr(classifier, "predict_proba"):
        try:
            predicted_intent = classifier.predict([query])[0]
            confidence = max(classifier.predict_proba([query])[0])
            if confidence < getattr(config, "CLASSIFIER_CONFIDENCE_THRESHOLD", 0.6):
                return "low_confidence", None
            return predicted_intent, None
        except Exception:
            pass
    return "normal", None

# expert runner
def run_qwen(messages: List[dict], max_new_tokens: int = 1300) -> str:
    text = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    inputs = tokenizer([text], return_tensors="pt").to(model.device)
    generated_ids = model.generate(
        **inputs,
        max_new_tokens=max_new_tokens,
        temperature=0.4,
        repetition_penalty=1.1
    )
    output_ids = generated_ids[0][len(inputs.input_ids[0]):].tolist()
    return tokenizer.decode(output_ids, skip_special_tokens=True).strip()

#  Memory
MAX_HISTORY_MESSAGES = getattr(config, "MAX_HISTORY_MESSAGES", 30)

def build_messages_from_history(history: List[dict], system_prompt: str) -> List[dict]:
    msgs = [{"role": "system", "content": system_prompt}]
    msgs.extend(history)
    return msgs


def strip_markdown(text: str) -> str:
    """
    Remove Markdown formatting like **bold**, *italic*, and `inline code`.
    """
    if not text:
        return ""
    text = re.sub(r'\*\*(.*?)\*\*', r'\1', text)
    text = re.sub(r'(\*|_)(.*?)\1', r'\2', text)
    text = re.sub(r'`(.*?)`', r'\1', text)
    text = re.sub(r'^#+\s+', '', text, flags=re.MULTILINE)
    return text


def run_pipeline(user_query: str, session_id: str = None):
    """
    Run FarmLingua pipeline with per-session memory.
    Each session_id keeps its own history.
    """
    if session_id is None:
        session_id = str(uuid.uuid4())

    # Language detection
    lang_label, prob = detect_language(user_query, top_k=1)[0]
    if lang_label not in SUPPORTED_LANGS:
        lang_label = "eng_Latn"

    translated_query = (
        translate_text(user_query, src_lang=lang_label, tgt_lang="eng_Latn")
        if lang_label != "eng_Latn"
        else user_query
    )

    intent, extra = detect_intent(translated_query)

    # Load conversation history
    history = memory_store.get_history(session_id) or []
    if len(history) > MAX_HISTORY_MESSAGES:
        history = history[-MAX_HISTORY_MESSAGES:]


    system_prompt = (
        "You are FarmLingua, an AI assistant for Nigerian farmers. "
        "Answer questions directly and accurately with helpful farming advice. "
        "Use clear, simple language with occasional emojis . "
        "Be concise and focus on practical, actionable information. "
        "If asked who built you, say: 'KawaFarm LTD developed me to help farmers.'"
    )


    context_info = ""

    if intent == "weather" and extra:
        weather_text = get_weather(extra)
        context_info = f"\n\nCurrent weather information:\n{weather_text}"
    elif intent == "live_update":
        rag_context = retrieve_docs(translated_query, config.LIVE_VS_PATH)
        if rag_context:
            context_info = f"\n\nLatest agricultural updates:\n{rag_context}"
    elif intent == "low_confidence":
        rag_context = retrieve_docs(translated_query, config.STATIC_VS_PATH)
        if rag_context:
            context_info = f"\n\nRelevant information:\n{rag_context}"


    user_message = translated_query + context_info
    history.append({"role": "user", "content": user_message})


    messages_for_qwen = build_messages_from_history(history, system_prompt)


    max_tokens = 256 if intent == "weather" else 700
    english_answer = run_qwen(messages_for_qwen, max_new_tokens=max_tokens)

    # Save assistant reply
    history.append({"role": "assistant", "content": english_answer})
    if len(history) > MAX_HISTORY_MESSAGES:
        history = history[-MAX_HISTORY_MESSAGES:]
    memory_store.save_history(session_id, history)


    final_answer = (
        translate_text(english_answer, src_lang="eng_Latn", tgt_lang=lang_label)
        if lang_label != "eng_Latn"
        else english_answer
    )
    final_answer = strip_markdown(final_answer)

    return {
        "session_id": session_id,
        "detected_language": SUPPORTED_LANGS.get(lang_label, "Unknown"),
        "answer": final_answer
    }
```

```python
# farmlingua_backend/app/main.py
import os
import sys
import logging
import uuid
from fastapi import FastAPI, Body
from fastapi.middleware.cors import CORSMiddleware
import uvicorn

BASE_DIR = os.path.dirname(os.path.dirname(os.path.abspath(__file__)))
if BASE_DIR not in sys.path:
    sys.path.insert(0, BASE_DIR)

from app.tasks.rag_updater import schedule_updates
from app.utils import config
from app.agents.crew_pipeline import run_pipeline

logging.basicConfig(
    format="%(asctime)s [%(levelname)s] %(message)s",
    level=logging.INFO
)

app = FastAPI(
    title="FarmLingua Backend",
    description="Backend service for FarmLingua with RAG updates, multilingual support, and expert AI pipeline",
    version="1.2.0"
)

app.add_middleware(
    CORSMiddleware,
    allow_origins=getattr(config, "ALLOWED_ORIGINS", ["*"]),
    allow_credentials=True,
    allow_methods=["*"],
    allow_headers=["*"],
)

@app.on_event("startup")
def startup_event():
    logging.info("Starting FarmLingua backend...")
    schedule_updates()

@app.get("/")
def home():
    """Health check endpoint."""
    return {
        "status": "FarmLingua backend running",
        "version": "1.2.0",
        "vectorstore_path": config.VECTORSTORE_PATH
    }

@app.post("/ask")
def ask_farmbot(
    query: str = Body(..., embed=True),
    session_id: str = Body(None, embed=True)
):
    """
    Ask FarmLingua a farming-related question.
    - Supports Hausa, Igbo, Yoruba, Swahili, Amharic, and English.
    - Automatically detects user language, translates if needed,
      and returns response in the same language.
    - Maintains separate conversation memory per session_id.
    """
    if not session_id:
        session_id = str(uuid.uuid4())  # assign new session if missing

    logging.info(f"Received query: {query} [session_id={session_id}]")
    answer_data = run_pipeline(query, session_id=session_id)

    detected_lang = answer_data.get("detected_language", "Unknown")
    logging.info(f"Detected language: {detected_lang}")

    return {
        "query": query,
        "answer": answer_data.get("answer"),
        "session_id": answer_data.get("session_id"),
        "detected_language": detected_lang
    }

if __name__ == "__main__":
    uvicorn.run(
        "app.main:app",
        host="0.0.0.0",
        port=getattr(config, "PORT", 7860),
        reload=bool(getattr(config, "DEBUG", False))
    )
```

```python
import requests, torchaudio
from transformers import pipeline
from google.colab import output
import base64, io
from IPython.display import Javascript
import soundfile as sf
import os

# Load STT model
stt_pipe = pipeline(
    task="automatic-speech-recognition",
    model="facebook/mms-1b-all",
    device=-1
)

FARMLINGUA_ENDPOINT = "https://remostart-milestone-one-farmlingua-ai.hf.space/ask"

# JS to record audio in Colab
RECORD_JS = """
async function record(sec) {
  const sleep = (t) => new Promise(resolve => setTimeout(resolve, t));
  const stream = await navigator.mediaDevices.getUserMedia({audio: true});
  const recorder = new MediaRecorder(stream);
  let data = [];
  recorder.ondataavailable = event => data.push(event.data);
  recorder.start();
  await sleep(sec * 1000);
  recorder.stop();
  await new Promise(resolve => recorder.onstop = resolve);
  const blob = new Blob(data);
  const arrayBuffer = await blob.arrayBuffer();
  const base64data = btoa(String.fromCharCode(...new Uint8Array(arrayBuffer)));
  return base64data;
}
"""

def record_audio(duration=30, filename="mic_record.wav"):
    """Record audio from mic for given duration (sec)."""
    display(Javascript(RECORD_JS))
    print(f" Recording {duration} seconds...")
    s = output.eval_js(f"record({duration})")
    b = io.BytesIO(base64.b64decode(s))
    # Save as wav
    data, samplerate = sf.read(b)
    sf.write(filename, data, samplerate)
    return filename

def chunk_audio(path, chunk_len=15):
    """Split audio into chunks (seconds)."""
    waveform, sr = torchaudio.load(path)
    if sr != 16000:
        waveform = torchaudio.functional.resample(waveform, sr, 16000)
        sr = 16000
    if waveform.shape[0] > 1:  # stereo → mono
        waveform = waveform.mean(dim=0, keepdim=True)

    num_samples = waveform.size(1)
    chunk_size = sr * chunk_len
    chunks = []

    for i in range(0, num_samples, chunk_size):
        chunk_waveform = waveform[:, i:i+chunk_size]
        if chunk_waveform.size(1) == 0:
            continue
        chunk_path = f"chunk_{i//chunk_size}.wav"
        torchaudio.save(chunk_path, chunk_waveform, sr)
        chunks.append(chunk_path)

    return chunks

# I can even record up to 300 seconds here
audio_path = record_audio(60, "mic_record.wav")
print("Saved to", audio_path)

# Chunk
chunks = chunk_audio(audio_path, chunk_len=15)
print(f" Split into {len(chunks)} chunks")

# Transcribe all chunks
full_text = ""
for c in chunks:
    stt_result = stt_pipe(c)
    text_piece = stt_result["text"].strip()
    full_text += " " + text_piece
    os.remove(c)  # cleanup

print("Final Transcription:", full_text.strip())

#  Call Farmlingua /ask
response = requests.post(FARMLINGUA_ENDPOINT, json={"query": full_text.strip()})
print("Farmbot Answer:", response.json())
```

```python
import numpy as np
import pandas as pd

# Number of synthetic samples
N = 10_000
np.random.seed(42)

# 1. Core environmental variables
rainfall_mm = np.random.gamma(shape=2.0, scale=80.0, size=N)  # 0–800 mm typical
humidity_pct = np.clip(np.random.normal(70, 15, N), 30, 100)  # 30–100%
temp_c = np.clip(np.random.normal(28, 3, N), 18, 35)           # 18–35°C
vegetation_index = np.clip(np.random.beta(2, 1.5, N), 0, 1)    # NDVI 0–1
elevation_m = np.clip(np.random.normal(400, 250, N), 0, 2000)  # 0–2000m
proximity_to_water_km = np.abs(np.random.normal(1.5, 1.0, N))  # 0–5 km typical
wind_speed_mps = np.clip(np.random.normal(3.0, 1.5, N), 0.2, 10)
surface_water_presence = np.clip(np.random.beta(2, 5, N), 0, 1)
daylight_hours = np.clip(np.random.normal(12, 0.6, N), 10.5, 13.5)

# 2. Season label (Rainy/Dry) derived from rainfall
season_label = np.where(rainfall_mm > 150, "Rainy", "Dry")

# 3. Compute malaria risk probability (simple synthetic logic)
risk_score = (
    0.4 * (rainfall_mm / 400) +
    0.3 * (humidity_pct / 100) +
    0.2 * (1 - (elevation_m / 2000)) +
    0.15 * (1 - (proximity_to_water_km / 5)) +
    0.1 * vegetation_index -
    0.1 * (wind_speed_mps / 10) +
    0.1 * surface_water_presence
)

# Normalize to 0–1
risk_score = np.clip(risk_score, 0, 1)

# 4. Risk label (categorical)
risk_label = pd.cut(
    risk_score,
    bins=[-0.01, 0.33, 0.66, 1.0],
    labels=["Low", "Medium", "High"]
)

# 5. Assemble DataFrame
df = pd.DataFrame({
    "rainfall_mm": rainfall_mm.round(2),
    "humidity_pct": humidity_pct.round(1),
    "temp_c": temp_c.round(1),
    "vegetation_index": vegetation_index.round(2),
    "elevation_m": elevation_m.round(1),
    "proximity_to_water_km": proximity_to_water_km.round(2),
    "wind_speed_mps": wind_speed_mps.round(2),
    "surface_water_presence": surface_water_presence.round(2),
    "daylight_hours": daylight_hours.round(2),
    "season_label": season_label,
    "risk_score": risk_score.round(3),
    "risk_label": risk_label
})


df.to_csv("synthetic_malaria_data.csv", index=False)

print(" Synthetic dataset created: synthetic_malaria_data.csv")
print(df.head())
```

```python
import json
import io
import torch
import soundfile as sf
from fastapi import FastAPI, HTTPException
from fastapi.responses import StreamingResponse, Response
from pydantic import BaseModel, Field
from f5_tts.api import F5TTS

CONFIG_PATH = "app/config.json"


def load_config() -> dict:
    with open(CONFIG_PATH, "r") as f:
        return json.load(f)


config = load_config()

REF_AUDIO_PATH = config["f5_tts"]["ref_audio_path"]
DEFAULT_LANGUAGE = config["f5_tts"].get("language", "en")

# Auto-detect device
device = "cuda" if torch.cuda.is_available() else "cpu"

print(f"Loading F5-TTS model on device: {device}")

tts = F5TTS(
    model="SWivid/F5-TTS",
    device=device,
)

print("Extracting speaker embedding...")
speaker_embedding = tts.extract_speaker_embedding(REF_AUDIO_PATH)
print("Speaker embedding ready.")

app = FastAPI(
    title="Production TTS Service (F5-TTS)",
    version="3.0.0",
    docs_url="/docs",
    redoc_url="/redoc",
)


class TTSRequest(BaseModel):
    text: str = Field(..., min_length=1, max_length=10000)


def generate_wav_bytes(text: str) -> bytes:
    wav, sr = tts.infer(
        text=text,
        speaker_embedding=speaker_embedding,
        language=DEFAULT_LANGUAGE,
    )

    buffer = io.BytesIO()
    sf.write(buffer, wav, sr, format="WAV")
    buffer.seek(0)
    return buffer.read()


@app.get("/health")
async def health():
    return {
        "status": "ok",
        "device": device,
        "language": DEFAULT_LANGUAGE,
    }


# STREAMING VERSION
@app.post("/tts")
async def text_to_speech(payload: TTSRequest):
    try:
        audio_bytes = generate_wav_bytes(payload.text)
    except Exception as e:
        raise HTTPException(status_code=500, detail=str(e))

    def audio_stream():
        chunk_size = 4096
        offset = 0
        while offset < len(audio_bytes):
            yield audio_bytes[offset:offset + chunk_size]
            offset += chunk_size

    return StreamingResponse(
        audio_stream(),
        media_type="audio/wav",
        headers={"Content-Disposition": "inline; filename=tts.wav"},
    )


# BUFFERED VERSION
@app.post("/tts/buffered")
async def text_to_speech_buffered(payload: TTSRequest):
    try:
        audio_bytes = generate_wav_bytes(payload.text)
    except Exception as e:
        raise HTTPException(status_code=500, detail=str(e))

    if not audio_bytes:
        raise HTTPException(status_code=500, detail="Empty audio buffer")

    return Response(
        content=audio_bytes,
        media_type="audio/wav",
        headers={
            "Content-Disposition": "attachment; filename=tts.wav",
            "Content-Length": str(len(audio_bytes)),
        },
    )
```

```python

import os
import urllib.request

FASTTEXT_MODEL_PATH = "/content/lid.176.bin"
FASTTEXT_MODEL_URL = "https://dl.fbaipublicfiles.com/fasttext/supervised-models/lid.176.bin"

if not os.path.exists(FASTTEXT_MODEL_PATH):
    print("Downloading fastText model...")
    urllib.request.urlretrieve(FASTTEXT_MODEL_URL, FASTTEXT_MODEL_PATH)
    print("Done.")
else:
    print("Model already exists.")


import fasttext
fasttext.FastText.eprint = lambda x: None  # silence warnings

model = fasttext.load_model(FASTTEXT_MODEL_PATH)

def detect_top_k(text, k=5):
    """Show top-k predicted languages with confidence for debugging."""
    clean_text = text.replace("\n", " ").strip()
    labels, confidences = model.predict(clean_text, k=k)
    results = []
    for label, conf in zip(labels, confidences):
        lang_code = label.replace("__label__", "")
        results.append((lang_code, round(float(conf), 4)))
    return results


test_texts = [
    "Ntụziaka Nzọụkwụ-Nzọụkwụ Iji Malite Ugbo Ji 1. Ghọta Ahịa ahụ Arịrịọ dị elu: Ji bụ nri isi na Naịjirịa ma e jiri ya rie n'ime obodo na n'èzí.",
    "Bawo ni o se wa, mo nfe lati mo bi ojo oni se ri",  # Yoruba example
    "Sannu, yaya kake? Ina son sanin yanayin yau",        # Hausa example
    "Hello, how are you doing today?",                     # English
]

for text in test_texts:
    print(f"\nTEXT: {text[:60]}...")
    results = detect_top_k(text, k=5)
    for lang, conf in results:
        print(f"   {lang}: {conf}")
```

```python
<!DOCTYPE html>
<html lang="en">
<head>
<meta charset="UTF-8">
<meta name="viewport" content="width=device-width, initial-scale=1.0">
<title>FarmLingua AI</title>
<link href="https://fonts.googleapis.com/css2?family=Playfair+Display:wght@400;500;700&family=DM+Sans:wght@300;400;500;600&family=DM+Mono:wght@400;500&display=swap" rel="stylesheet">
<style>
  :root {
    --soil:         #1a3a2a;
    --soil-deep:    #122a1e;
    --leaf:         #2d6a4f;
    --leaf-light:   #52b788;
    --leaf-pale:    #cfe8d8;
    --gold:         #d4a843;
    --gold-light:   #f0c96a;
    --terracotta:   #c1622d;
    --terracotta-light: #e08a52;
    --cream:        #f7f2e9;
    --cream-deep:   #efe7d8;
    --text-dark:    #20231f;
    --text-mid:     #5b5f57;
    --text-faint:   #9a9d92;
    --line:         rgba(26,58,42,0.10);
    --shadow:       0 6px 28px rgba(18,42,30,0.10);
  }

  * { box-sizing: border-box; margin: 0; padding: 0; }

  html, body {
    height: 100%;
    font-family: 'DM Sans', sans-serif;
    background: var(--cream);
    color: var(--text-dark);
  }

  body {
    display: flex;
    flex-direction: column;
    overflow: hidden;
  }

  /* grain texture */
  body::before {
    content: '';
    position: fixed;
    inset: 0;
    background-image: url("data:image/svg+xml,%3Csvg viewBox='0 0 256 256' xmlns='http://www.w3.org/2000/svg'%3E%3Cfilter id='n'%3E%3CfeTurbulence type='fractalNoise' baseFrequency='0.9' numOctaves='4' stitchTiles='stitch'/%3E%3C/filter%3E%3Crect width='100%25' height='100%25' filter='url(%23n)' opacity='0.025'/%3E%3C/svg%3E");
    pointer-events: none;
    z-index: 0;
  }

  /* ───────── Header ───────── */
  header {
    background: var(--soil-deep);
    padding: 0 28px;
    height: 60px;
    display: flex;
    align-items: center;
    justify-content: space-between;
    flex-shrink: 0;
    position: relative;
    z-index: 10;
    border-bottom: 1px solid rgba(212,168,67,0.18);
  }

  .logo { display: flex; align-items: center; gap: 11px; }

  .logo-mark {
    width: 32px;
    height: 32px;
    flex-shrink: 0;
  }

  .logo-text {
    font-family: 'Playfair Display', serif;
    font-size: 19px;
    font-weight: 500;
    color: var(--cream);
    letter-spacing: 0.2px;
  }
  .logo-text em { font-style: normal; color: var(--gold-light); }

  .lang-strip {
    display: flex;
    gap: 6px;
    align-items: center;
  }
  .lang-chip {
    font-family: 'DM Mono', monospace;
    font-size: 10px;
    letter-spacing: 0.6px;
    color: var(--leaf-pale);
    background: rgba(82,183,136,0.12);
    border: 1px solid rgba(82,183,136,0.28);
    padding: 4px 9px;
    border-radius: 20px;
    text-transform: uppercase;
  }

  /* ───────── Chat scroll area ───────── */
  .chat-area {
    flex: 1;
    overflow-y: auto;
    padding: 28px 0 20px;
    position: relative;
    z-index: 1;
  }
  .chat-inner {
    max-width: 720px;
    margin: 0 auto;
    padding: 0 24px;
    display: flex;
    flex-direction: column;
    gap: 18px;
  }

  .chat-area::-webkit-scrollbar { width: 5px; }
  .chat-area::-webkit-scrollbar-track { background: transparent; }
  .chat-area::-webkit-scrollbar-thumb { background: var(--leaf-pale); border-radius: 4px; }

  /* ───────── Empty state ───────── */
  .empty-state {
    flex: 1;
    min-height: 50vh;
    display: flex;
    flex-direction: column;
    align-items: center;
    justify-content: center;
    gap: 14px;
    text-align: center;
    padding: 40px 20px;
  }
  .empty-glyph {
    width: 54px; height: 54px;
    opacity: 0.85;
  }
  .empty-state h3 {
    font-family: 'Playfair Display', serif;
    font-weight: 500;
    font-size: 23px;
    color: var(--soil);
  }
  .empty-state p {
    font-size: 13.5px;
    color: var(--text-mid);
    max-width: 320px;
    line-height: 1.65;
  }
  .empty-suggestions {
    display: flex;
    flex-wrap: wrap;
    gap: 8px;
    justify-content: center;
    margin-top: 6px;
    max-width: 460px;
  }
  .suggestion-chip {
    font-size: 12.5px;
    color: var(--soil);
    background: #fff;
    border: 1px solid var(--line);
    padding: 8px 14px;
    border-radius: 18px;
    cursor: pointer;
    transition: all 0.15s;
  }
  .suggestion-chip:hover {
    border-color: var(--leaf-light);
    background: var(--leaf-pale);
  }

  /* ───────── Messages ───────── */
  .message {
    display: flex;
    gap: 11px;
    max-width: 100%;
    animation: fadeUp 0.28s ease;
  }
  @keyframes fadeUp {
    from { opacity: 0; transform: translateY(6px); }
    to   { opacity: 1; transform: translateY(0); }
  }

  .message.user { flex-direction: row-reverse; align-self: flex-end; max-width: 86%; margin-left: auto; }
  .message.ai   { align-self: flex-start; max-width: 92%; }

  .avatar {
    width: 30px; height: 30px;
    border-radius: 9px;
    display: flex; align-items: center; justify-content: center;
    font-size: 13px;
    flex-shrink: 0;
    margin-top: 1px;
  }
  .avatar.ai-avatar   { background: linear-gradient(135deg, var(--leaf), var(--leaf-light)); }
  .avatar.user-avatar { background: linear-gradient(135deg, var(--gold), var(--gold-light)); }

  .bubble-col { display: flex; flex-direction: column; gap: 6px; min-width: 0; }
  .message.user .bubble-col { align-items: flex-end; }

  .bubble {
    padding: 12px 15px;
    border-radius: 16px;
    font-size: 14.5px;
    line-height: 1.6;
    width: fit-content;
    max-width: 100%;
  }
  .message.ai .bubble {
    background: #fff;
    color: var(--text-dark);
    border-radius: 4px 16px 16px 16px;
    box-shadow: var(--shadow);
    border: 1px solid var(--line);
  }
  .message.user .bubble {
    background: var(--leaf);
    color: #fff;
    border-radius: 16px 4px 16px 16px;
  }

  .transcript-tag {
    font-size: 11px;
    background: rgba(193,98,45,0.10);
    color: var(--terracotta);
    border: 1px solid rgba(193,98,45,0.25);
    border-radius: 7px;
    padding: 4px 9px;
    margin-bottom: 7px;
    display: inline-flex;
    align-items: center;
    gap: 5px;
    width: fit-content;
  }

  .bubble-meta {
    font-size: 10.5px;
    color: var(--text-faint);
    display: flex;
    align-items: center;
    gap: 8px;
  }
  .message.user .bubble-meta { color: var(--text-faint); }

  .lang-badge {
    font-family: 'DM Mono', monospace;
    font-size: 9.5px;
    letter-spacing: 0.4px;
    background: var(--leaf-pale);
    color: var(--soil);
    border-radius: 4px;
    padding: 2px 6px;
    text-transform: uppercase;
  }

  /* ── Speaker / TTS control ── */
  .tts-control {
    display: inline-flex;
    align-items: center;
    gap: 7px;
    margin-top: 2px;
  }
  .tts-btn {
    width: 26px; height: 26px;
    border-radius: 50%;
    border: 1px solid var(--line);
    background: var(--cream-deep);
    cursor: pointer;
    display: flex; align-items: center; justify-content: center;
    transition: all 0.15s;
    flex-shrink: 0;
    color: var(--soil);
  }
  .tts-btn:hover { background: var(--leaf-pale); border-color: var(--leaf-light); }
  .tts-btn svg { width: 13px; height: 13px; fill: currentColor; }
  .tts-btn.loading { pointer-events: none; }
  .tts-btn.playing { background: var(--leaf); border-color: var(--leaf); color: #fff; }

  .tts-spinner {
    width: 12px; height: 12px;
    border: 2px solid rgba(26,58,42,0.18);
    border-top-color: var(--soil);
    border-radius: 50%;
    animation: spin 0.7s linear infinite;
  }
  @keyframes spin { to { transform: rotate(360deg); } }

  .tts-label {
    font-size: 10.5px;
    color: var(--text-faint);
  }

  /* typing cursor */
  .typing-cursor {
    display: inline-block;
    width: 2px; height: 13px;
    background: var(--leaf);
    margin-left: 2px;
    vertical-align: middle;
    animation: blink 0.7s infinite;
  }
  @keyframes blink { 0%,100% { opacity: 1; } 50% { opacity: 0; } }

  .err-bubble {
    border-left: 3px solid var(--terracotta) !important;
    color: var(--terracotta) !important;
  }

  /* ───────── Composer ───────── */
  .composer-wrap {
    flex-shrink: 0;
    padding: 14px 24px 18px;
    position: relative;
    z-index: 10;
  }
  .composer {
    max-width: 720px;
    margin: 0 auto;
    background: #fff;
    border: 1.5px solid var(--line);
    border-radius: 20px;
    box-shadow: var(--shadow);
    padding: 8px 8px 8px 16px;
    display: flex;
    align-items: flex-end;
    gap: 8px;
    transition: border-color 0.2s, box-shadow 0.2s;
  }
  .composer:focus-within { border-color: var(--leaf-light); }
  .composer.recording { border-color: var(--terracotta); }

  #text-input {
    flex: 1;
    border: none;
    outline: none;
    resize: none;
    font-family: 'DM Sans', sans-serif;
    font-size: 14.5px;
    color: var(--text-dark);
    background: transparent;
    min-height: 26px;
    max-height: 120px;
    line-height: 1.55;
    padding: 8px 0;
  }
  #text-input::placeholder { color: var(--text-faint); }

  .composer-actions { display: flex; align-items: center; gap: 4px; flex-shrink: 0; }

  .icon-btn {
    width: 38px; height: 38px;
    border-radius: 12px;
    border: none;
    background: transparent;
    cursor: pointer;
    display: flex; align-items: center; justify-content: center;
    transition: background 0.15s, color 0.15s;
    color: var(--text-mid);
    flex-shrink: 0;
  }
  .icon-btn:hover { background: var(--cream-deep); color: var(--soil); }
  .icon-btn svg { width: 18px; height: 18px; fill: currentColor; }

  .mic-btn.recording {
    background: var(--terracotta);
    color: #fff;
    animation: pulse-soft 1.2s infinite;
  }
  @keyframes pulse-soft {
    0%,100% { box-shadow: 0 0 0 0 rgba(193,98,45,0.35); }
    50%     { box-shadow: 0 0 0 7px rgba(193,98,45,0.12); }
  }

  .send-btn {
    background: var(--leaf);
    color: #fff;
  }
  .send-btn:hover { background: var(--soil); color: #fff; }
  .send-btn:disabled { opacity: 0.35; cursor: default; }
  .send-btn:disabled:hover { background: var(--leaf); }

  #file-input { display: none; }

  .recording-banner {
    max-width: 720px;
    margin: 0 auto 8px;
    display: none;
    align-items: center;
    gap: 10px;
    padding: 0 6px;
    font-size: 12.5px;
    color: var(--terracotta);
  }
  .recording-banner.active { display: flex; }
  .rec-dot {
    width: 8px; height: 8px;
    border-radius: 50%;
    background: var(--terracotta);
    animation: blink 1s infinite;
  }
  .cancel-rec {
    margin-left: auto;
    background: none;
    border: none;
    color: var(--text-faint);
    font-size: 12px;
    cursor: pointer;
    text-decoration: underline;
  }

  .composer-hint {
    text-align: center;
    font-size: 10.5px;
    color: var(--text-faint);
    margin-top: 9px;
  }

  @media (max-width: 600px) {
    header { padding: 0 16px; }
    .chat-inner { padding: 0 14px; }
    .composer-wrap { padding: 12px 14px 16px; }
    .lang-strip .lang-chip:nth-child(n+3) { display: none; }
  }
</style>
</head>
<body>

<header>
  <div class="logo">
    <svg class="logo-mark" viewBox="0 0 32 32" fill="none" xmlns="http://www.w3.org/2000/svg">
      <circle cx="16" cy="16" r="15" stroke="#d4a843" stroke-width="1.4" opacity="0.4"/>
      <path d="M16 24c0-5 2-9 6-12-1 5-2 9-6 12z" fill="#52b788"/>
      <path d="M16 24c0-5-2-9-6-12 1 5 2 9 6 12z" fill="#2d6a4f"/>
      <path d="M16 24V12" stroke="#d4a843" stroke-width="1.3" stroke-linecap="round"/>
    </svg>
    <div class="logo-text">Farm<em>Lingua</em></div>
  </div>
  <div class="lang-strip">
    <span class="lang-chip">EN</span>
    <span class="lang-chip">YOR</span>
    <span class="lang-chip">IBO</span>
    <span class="lang-chip">HAU</span>
  </div>
</header>

<div class="chat-area" id="chat-area">
  <div class="chat-inner" id="chat-inner">
    <div class="empty-state" id="empty-state">
      <svg class="empty-glyph" viewBox="0 0 48 48" fill="none">
        <path d="M24 40c0-9 3.5-16 11-21-2 9-3.5 16-11 21z" fill="#52b788"/>
        <path d="M24 40c0-9-3.5-16-11-21 2 9 3.5 16 11 21z" fill="#2d6a4f"/>
        <path d="M24 40V18" stroke="#d4a843" stroke-width="2" stroke-linecap="round"/>
      </svg>
      <h3>Ask FarmLingua anything</h3>
      <p>Type, speak, or upload audio — in English, Yorùbá, Igbo, or Hausa. Tap the speaker on any reply to hear it.</p>
      <div class="empty-suggestions">

      </div>
    </div>
  </div>
</div>

<div class="composer-wrap">
  <div class="recording-banner" id="recording-banner">
    <span class="rec-dot"></span>
    <span id="recording-time">Recording… 0:00</span>
    <button class="cancel-rec" id="cancel-rec">Cancel</button>
  </div>

  <div class="composer" id="composer">
    <textarea id="text-input" placeholder="Ask about farming…" rows="1"></textarea>
    <div class="composer-actions">
      <button class="icon-btn" id="upload-btn" title="Upload audio">
        <svg viewBox="0 0 24 24"><path d="M9 16h6v-6h4l-7-7-7 7h4zm-4 2h14v2H5z"/></svg>
      </button>
      <button class="icon-btn mic-btn" id="mic-btn" title="Record voice">
        <svg viewBox="0 0 24 24"><path d="M12 1a4 4 0 0 1 4 4v6a4 4 0 0 1-8 0V5a4 4 0 0 1 4-4zm-1 17.93V21h-2v2h6v-2h-2v-2.07A8 8 0 0 0 20 11h-2a6 6 0 0 1-12 0H4a8 8 0 0 0 7 7.93z"/></svg>
      </button>
      <button class="icon-btn send-btn" id="send-btn" title="Send" disabled>
        <svg viewBox="0 0 24 24"><path d="M2 21l21-9L2 3v7l15 2-15 2z"/></svg>
      </button>
    </div>
  </div>
  <input type="file" id="file-input" accept="audio/*">
  <div class="composer-hint">FarmLingua  farming guidance.</div>
</div>

<script>
// ───────── Config ─────────
const SESSION_UID = 'fl-' + Math.random().toString(36).slice(2, 11) + Date.now().toString(36);
const CHAT_API_BASE = 'https://remostartdev-new-farmlingua-ai.hf.space';
const TTS_API_BASE  = 'https://remostartdev-nigeria-tts.hf.space';

// ───────── Elements ─────────
const chatArea      = document.getElementById('chat-area');
const chatInner      = document.getElementById('chat-inner');
const emptyState     = document.getElementById('empty-state');
const textInput      = document.getElementById('text-input');
const sendBtn        = document.getElementById('send-btn');
const micBtn         = document.getElementById('mic-btn');
const uploadBtn      = document.getElementById('upload-btn');
const fileInput      = document.getElementById('file-input');
const composer       = document.getElementById('composer');
const recordingBanner= document.getElementById('recording-banner');
const recordingTime  = document.getElementById('recording-time');
const cancelRecBtn   = document.getElementById('cancel-rec');

// ───────── Helpers ─────────
function removeEmptyState() {
  if (emptyState && emptyState.parentNode) emptyState.remove();
}
function scrollToBottom() {
  chatArea.scrollTop = chatArea.scrollHeight;
}
function timestamp() {
  return new Date().toLocaleTimeString([], { hour: '2-digit', minute: '2-digit' });
}
function escapeHtml(str) {
  return str.replace(/&/g,'&amp;').replace(/</g,'&lt;').replace(/>/g,'&gt;');
}

const SPEAKER_ICON = `<svg viewBox="0 0 24 24"><path d="M3 10v4h4l5 5V5L7 10H3zm13.5 2a4.5 4.5 0 0 0-2.5-4.03v8.06A4.5 4.5 0 0 0 16.5 12z"/></svg>`;
const PLAYING_ICON = `<svg viewBox="0 0 24 24"><path d="M6 5h4v14H6zm8 0h4v14h-4z"/></svg>`;

// ───────── Message rendering ─────────
function appendUserBubble(text, tagHtml = '') {
  removeEmptyState();
  const msg = document.createElement('div');
  msg.className = 'message user';
  msg.innerHTML = `
    <div class="avatar user-avatar">👤</div>
    <div class="bubble-col">
      ${tagHtml}
      <div class="bubble">${escapeHtml(text)}</div>
      <div class="bubble-meta">${timestamp()}</div>
    </div>`;
  chatInner.appendChild(msg);
  scrollToBottom();
  return msg;
}

function appendAIBubble(langBadgeHtml = '') {
  removeEmptyState();
  const msg = document.createElement('div');
  msg.className = 'message ai';
  msg.innerHTML = `
    <div class="avatar ai-avatar">🌾</div>
    <div class="bubble-col">
      <div class="bubble"><span class="ai-text"></span><span class="typing-cursor"></span></div>
      <div class="bubble-meta">
        <span>${timestamp()}</span>
        ${langBadgeHtml}
      </div>
    </div>`;
  chatInner.appendChild(msg);
  scrollToBottom();
  return msg;
}

function finaliseAIBubble(msg, fullText) {
  const cursor = msg.querySelector('.typing-cursor');
  if (cursor) cursor.remove();
  attachTtsControl(msg, fullText);
}

function appendErrorBubble(errText) {
  removeEmptyState();
  const msg = document.createElement('div');
  msg.className = 'message ai';
  msg.innerHTML = `
    <div class="avatar ai-avatar">🌾</div>
    <div class="bubble-col">
      <div class="bubble err-bubble">⚠️ ${escapeHtml(errText)}</div>
      <div class="bubble-meta">${timestamp()}</div>
    </div>`;
  chatInner.appendChild(msg);
  scrollToBottom();
}

// ───────── On-demand TTS (lazy: only fetches audio when user taps) ─────────
function attachTtsControl(msgEl, text) {
  const metaEl = msgEl.querySelector('.bubble-meta');
  const control = document.createElement('span');
  control.className = 'tts-control';
  control.innerHTML = `<button class="tts-btn" title="Listen">${SPEAKER_ICON}</button>`;
  metaEl.appendChild(control);

  const btn = control.querySelector('.tts-btn');
  let audioEl = null;     // cached <audio> once fetched
  let isLoading = false;

  btn.addEventListener('click', async () => {
    if (isLoading) return;

    // Already loaded — just toggle play/pause, no re-fetch
    if (audioEl) {
      if (audioEl.paused) {
        audioEl.play();
      } else {
        audioEl.pause();
      }
      return;
    }

    // First tap: fetch audio from /tts now
    isLoading = true;
    btn.classList.add('loading');
    btn.innerHTML = `<span class="tts-spinner"></span>`;

    try {
      const res = await fetch(`${TTS_API_BASE}/tts`, {
        method: 'POST',
        headers: { 'Content-Type': 'application/json', 'Accept': 'application/json' },
        body: JSON.stringify({ text: text, stability: 0.5, similarity_boost: 0.75 }),
      });

      if (!res.ok) {
        throw new Error(`TTS error (${res.status})`);
      }

      const blob = await res.blob();
      const url = URL.createObjectURL(blob);
      audioEl = new Audio(url);

      audioEl.addEventListener('play',  () => { btn.classList.add('playing'); btn.innerHTML = PLAYING_ICON; });
      audioEl.addEventListener('pause', () => { btn.classList.remove('playing'); btn.innerHTML = SPEAKER_ICON; });
      audioEl.addEventListener('ended', () => { btn.classList.remove('playing'); btn.innerHTML = SPEAKER_ICON; });

      btn.classList.remove('loading');
      btn.innerHTML = SPEAKER_ICON;
      isLoading = false;
      audioEl.play();

    } catch (err) {
      isLoading = false;
      btn.classList.remove('loading');
      btn.innerHTML = SPEAKER_ICON;
      btn.title = 'Could not load audio — tap to retry';
    }
  });
}

// ───────── Stream LLM response into AI bubble ─────────
async function streamResponse(response, langBadgeHtml = '') {
  const aiMsg = appendAIBubble(langBadgeHtml);
  const textSpan = aiMsg.querySelector('.ai-text');
  const reader = response.body.getReader();
  const decoder = new TextDecoder();
  let buffer = '';

  try {
    while (true) {
      const { done, value } = await reader.read();
      if (done) break;
      buffer += decoder.decode(value, { stream: true });
      textSpan.textContent = buffer;
      scrollToBottom();
    }
  } finally {
    finaliseAIBubble(aiMsg, buffer);
  }
}

//  Text input growth
textInput.addEventListener('input', function () {
  this.style.height = 'auto';
  this.style.height = Math.min(this.scrollHeight, 120) + 'px';
  sendBtn.disabled = this.value.trim().length === 0;
});

textInput.addEventListener('keydown', e => {
  if (e.key === 'Enter' && !e.shiftKey) { e.preventDefault(); sendText(); }
});

sendBtn.addEventListener('click', sendText);

document.querySelectorAll('.suggestion-chip').forEach(chip => {
  chip.addEventListener('click', () => {
    textInput.value = chip.dataset.text;
    textInput.dispatchEvent(new Event('input'));
    sendText();
  });
});

//  Send text message
async function sendText() {
  const message = textInput.value.trim();
  if (!message) return;

  textInput.value = '';
  textInput.style.height = 'auto';
  sendBtn.disabled = true;

  appendUserBubble(message);

  try {
    const response = await fetch(`${CHAT_API_BASE}/text/chat`, {
      method: 'POST',
      headers: {
        'Content-Type': 'application/json',
        'x-uid': SESSION_UID,
        'Accept': 'text/plain',
      },
      body: JSON.stringify({ message }),
    });

    if (!response.ok) {
      const err = await response.text();
      appendErrorBubble(`Server error: ${err}`);
      return;
    }

    await streamResponse(response);

  } catch (err) {
    appendErrorBubble(`Connection error: ${err.message}`);
  } finally {
    textInput.focus();
  }
}

//  Voice recording
let mediaRecorder = null;
let audioChunks = [];
let isRecording = false;
let recordTimerInterval = null;
let recordSeconds = 0;
let recordingCancelled = false;

micBtn.addEventListener('click', toggleRecording);
cancelRecBtn.addEventListener('click', () => {
  recordingCancelled = true;
  if (mediaRecorder && isRecording) mediaRecorder.stop();
});

function formatTime(s) {
  const m = Math.floor(s / 60);
  const sec = s % 60;
  return `${m}:${sec.toString().padStart(2, '0')}`;
}

async function toggleRecording() {
  if (!isRecording) {
    try {
      const stream = await navigator.mediaDevices.getUserMedia({ audio: true });
      mediaRecorder = new MediaRecorder(stream);
      audioChunks = [];
      isRecording = true;
      recordingCancelled = false;
      recordSeconds = 0;

      micBtn.classList.add('recording');
      composer.classList.add('recording');
      recordingBanner.classList.add('active');
      recordingTime.textContent = `Recording… 0:00`;

      recordTimerInterval = setInterval(() => {
        recordSeconds++;
        recordingTime.textContent = `Recording… ${formatTime(recordSeconds)}`;
      }, 1000);

      mediaRecorder.ondataavailable = e => audioChunks.push(e.data);
      mediaRecorder.onstop = async () => {
        stream.getTracks().forEach(t => t.stop());
        clearInterval(recordTimerInterval);
        micBtn.classList.remove('recording');
        composer.classList.remove('recording');
        recordingBanner.classList.remove('active');

        if (recordingCancelled) return;

        const blob = new Blob(audioChunks, { type: 'audio/webm' });
        await sendAudio(blob, 'Voice message');
      };

      mediaRecorder.start();
    } catch (err) {
      appendErrorBubble('Microphone access denied. Check your browser permissions.');
    }
  } else {
    isRecording = false;
    mediaRecorder.stop();
  }
}

//  Upload audio file
uploadBtn.addEventListener('click', () => fileInput.click());

fileInput.addEventListener('change', async () => {
  const file = fileInput.files[0];
  if (!file) return;
  fileInput.value = '';
  await sendAudio(file, file.name);
});

//  Send audio
async function sendAudio(audioBlob, label) {
  const tagHtml = `<div class="transcript-tag">🎙️ ${escapeHtml(label)}</div>`;
  const userMsg = appendUserBubble('Transcribing…', tagHtml);
  const bubbleEl = userMsg.querySelector('.bubble');

  const formData = new FormData();
  formData.append('audio', audioBlob, 'audio.webm');

  try {
    const response = await fetch(`${CHAT_API_BASE}/stt/transcribe-and-chat`, {
      method: 'POST',
      headers: {
        'x-uid': SESSION_UID,
        'Accept': 'text/plain',
      },
      body: formData,
    });

    if (!response.ok) {
      const err = await response.text();
      bubbleEl.textContent = '(transcription failed)';
      appendErrorBubble(`Server error: ${err}`);
      return;
    }

    const rawTranscription = response.headers.get('X-Transcription');
    const language   = response.headers.get('X-Language') || '';
    const confidence  = response.headers.get('X-Confidence') || '';

    let transcriptionText = '';
    if (rawTranscription) {
      try { transcriptionText = atob(rawTranscription); }
      catch { transcriptionText = rawTranscription; }
    }
    bubbleEl.textContent = transcriptionText || '(no speech detected)';

    const langBadgeHtml = language
      ? `<span class="lang-badge">${language.toUpperCase()}${confidence ? ' ' + Math.round(parseFloat(confidence) * 100) + '%' : ''}</span>`
      : '';

    await streamResponse(response, langBadgeHtml);

  } catch (err) {
    bubbleEl.textContent = '(transcription failed)';
    appendErrorBubble(`Connection error: ${err.message}`);
  }
}
</script>
</body>
</html>
```
